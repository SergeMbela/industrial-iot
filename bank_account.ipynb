Exemple de gestion ajout et retrait dans son compte

```python
class Bankaccount:
    def __init__(self, bank, titulaire, balance = 0.0):   
        self.bank = bank
        self.titulaire = titulaire
        self.balance  = balance
    #Ajout d'un montant dans son compte
    def add  (self,bank,amount): 
        if amount > 0:
            self.bank = bank
            self.balance = amount + self.balance
            print (self.balance)
        else:
            print("The amount must be positive")   

    #Retrait d'un montant dans son compte  
    #Il faut que le montant à retirer soit inférieur ou égal à celui qui est en compte    
    def retrait (self,bank,amount): 
        if amount <= self.balance:
            self.bank = bank
            self.balance = self.balance - amount
            print(self.balance)
        else:
            print("L_Vous ne disposez pas assez sur votre compte")

    ## Vérifier le solde
            print(f"Votre solde est de  {self.balance}")    

## Ma situation actuelle
account = Bankaccount("Bnp ParisBas","Serge",100)  
account.add("Bnp ParisBas",20)   
account.retrait("Bnp ParisBas",20)       
```
